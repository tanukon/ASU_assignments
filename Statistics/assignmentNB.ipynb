```python
import numpy
import numpy as np
import scipy.io
import math
import geneNewData

def main():
    myID='1234'
    geneNewData.geneData(myID)
    Numpyfile0 = scipy.io.loadmat('digit0_stu_train'+myID+'.mat')
    Numpyfile1 = scipy.io.loadmat('digit1_stu_train'+myID+'.mat')
    Numpyfile2 = scipy.io.loadmat('digit0_testset'+'.mat')
    Numpyfile3 = scipy.io.loadmat('digit1_testset'+'.mat')
    train0 = Numpyfile0.get('target_img')
    train1 = Numpyfile1.get('target_img')
    test0 = Numpyfile2.get('target_img')
    test1 = Numpyfile3.get('target_img')
    print([len(train0),len(train1),len(test0),len(test1)])
    print('Your trainset and testset are generated successfully!')
    
    
    # Task1
    # Students must write the code that calculates each imageâ€™s mean and variance.
    mean_train0 = np.mean(train0)
    mean_train1 = np.mean(train1)
    mean_test0 = np,mean(test0)
    mean_test1 = np.mean(test1)
    
    print(f"Each image mean: train0 {mean_train0}, train1 {mean_train1}, test0 {mean_test0}, test1 {mean_test1}")
    
    # Task1
    # Students must write a code that is able to calculate each image mean and variance.
    train_data = np.vstack([train0, train1])
    test_data = np.vstack([test0, test1])
    
    train_feature1 = np.mean(train_data.reshape(10000,-1),axis=1,keepdims=True)
    train_feature2 = np.std(train_data.reshape(10000,-1),axis=1,keepdims=True)
    
    test_feature1 = np.mean(test_data.reshape(test_data.shape[0], -1), axis=1, keepdims=True)
    test_feature2 = np.std(test_data.reshape(test_data.shape[0], -1), axis=1, keepdims=True)

    # Task2
    # Students need to calculate all the parameters for 2 class naive bayes classifiers
    mean_feature1_0 = np.mean(train_feature1[:5000])
    var_feature1_0 = np.var(train_feature1[:5000])
    mean_feature2_0 = np.mean(train_feature2[:5000])
    var_feature2_0 = np.var(train_feature2[:5000])
    
    mean_feature1_1 = np.mean(train_feature1[5000:])
    var_feature1_1 = np.var(train_feature1[5000:])
    mean_feature2_1 = np.mean(train_feature2[5000:])
    var_feature2_1 = np.var(train_feature2[5000:])
    
    # Task3
    # getting a result from naive baise classifier
    
    def calc_condition_prob(x, variance, mu):
        p = (1/(2*np.pi*variance)**0.5) * np.exp(-(x-mu)**2/(2*variance)) 
    
        return p
    
    feature1_0 = calc_condition_prob(test_feature1, var_feature1_0, mean_feature1_0)
    feature1_1 = calc_condition_prob(test_feature1, var_feature1_1, mean_feature1_1)
    feature2_0 = calc_condition_prob(test_feature2, var_feature2_0, mean_feature2_0)
    feature2_1 = calc_condition_prob(test_feature2, var_feature2_1, mean_feature2_1)
    
    predict_y_x0 = feature1_0 * feature2_0 * 0.5
    predict_y_x1 = feature1_1 * feature2_1 * 0.5
    
    p_digit0 = len(test_data[:len(test0)][np.where(predict_y_x0[:len(test0)] > predict_y_x1[:len(test0)])])/len(test0)
    p_digit1 = len(test_data[:len(test0)][np.where(predict_y_x1[:len(test0)] > predict_y_x0[:len(test0)])])/len(test1)
    
if __name__ == '__main__':
    main()
```
